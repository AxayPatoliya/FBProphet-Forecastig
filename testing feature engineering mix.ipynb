# Probability Ratio Encoding
1. Probability of Survived based on Cabin--- Categorical Feature
2. Probability of Not Survived---1-pr(Survived)
3. pr(Survived)/pr(Not Survived)
4. Dictonary to map cabin with probability
5. replace with the categorical feature

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv', usecols=['Cabin', 'Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
for i in df['Cabin']:
    print(i[0])
```

```python
def getFirst(i):
    return i[0]
```

```python
df['Cabin'] = df['Cabin'].apply(getFirst)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
# getting the survival rate based on the cabin column
prob_df = df.groupby(['Cabin'])['Survived'].mean()
print(prob_df)
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died'] = 1 - prob_df['Survived']
```

```python
prob_df
```

```python
# Probability Ration can be defined as the (probablity of survived)/(probability of died)
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df
```

```python
probablity_encoded = prob_df['Probability_ratio'].to_dict()
print(probablity_encoded)
```

```python
# aapda df ma je cabin column hatu aena aadhare aapde aek navu feature banaisu ke jema aapde cabin column ne map karisu ne with probablity_encoded 
df['Probability_encoded'] = df['Cabin'].map(probablity_encoded)
# nava feature ma cabin ma jya jya 'A' hase tya 0.875 aai jase ae rite probablity_encoded thi map karyu che aene according value aai jase
```
